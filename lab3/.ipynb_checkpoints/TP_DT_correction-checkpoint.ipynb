```python
import jax
import jax.numpy as jnp
import numpy as np
import gzip
import pickle
import matplotlib.pyplot as plt
import time
```

For this lab, we will use the bluebell dataset. It consists of $64\times 64$ color images, which we will have to flatten into $12k$ dimensional vectors. The code for the dataset comes with several train/val/test splits, but in this notebook, we will use the first split and do our own cross-validation routines.

```python
# Load the dataset
from bluebell import Bluebell
X_train_ds = Bluebell('bluebell_64', 'train', split=0)
X_val_ds = Bluebell('bluebell_64', 'val', split=0)
X_train = np.array([img.flatten()/127.5 - 1. for img, lab in X_train_ds])
y_train = np.array([lab for img, lab in X_train_ds])
X_val = np.array([img.flatten()/127.5 - 1. for img, lab in X_val_ds])
y_val = np.array([lab for img, lab in X_val_ds])
plt.imshow(X_train[0].reshape(64, 64, 3)/2+0.5)
print(y_train[0])
```

Next, we want to reduce the number of dimensions that will be search through with the decision trees. Since our images are $64\times 64\times 3$ values, this leads to a very high dimensional space that has to be searched at each step. However, dimensions where all images have the same value, or very close values, will never be selected in the tree because they do not provide a good gain.

We will thus select only the 2048 dimension with the highest variance to perform our analysis.

```python
dim = jnp.argsort(X_train.std(axis=0), descending=True)[0:2048]
X_train = jnp.array(X_train[:, dim])
X_val = jnp.array(X_val[:, dim])
print(X_train.shape, X_val.shape)
```

## Implementing a randomized Decision Tree

<div class="alert alert-success"> 
    <b>Q1.</b> Implement the code of a function that finds an optimal threshold along a given dimension, using the $0-1$ loss with specified example weights and test it on the 150th dimension. To speed-up things, we will only consider 8 thresholds between the minimum and maximum value (use 'linspace'). Compare it to assigning a unique label to all samples. You should get a significant decrease of loss from ~0.92 (random 1/12 chance) to ~0.84. <i>(Indicative time: 10 minutes for a slow version, but take the extra 20 minutes to have a parallel version testing all thresholds at once that runs in under 1s, it is worth it for the next questions.)</i>
</div>

```python
'''
takes arguements 
y_pred: prediction
y_true: true labels
weights: weights for each example
'''
@jax.jit
def zeroOneLoss(y_pred, y_true, weights):
    return (weights * (y_pred != y_true)).sum(axis=0)/(1e-12+weights.sum(axis=0))
```

```python
'''
takes arguments
X: training samples
y: training labels
dim: dimension to use
w: weight associated to each example (can be True/False or 1/0 to remove some examples)
returns the gain and the threshold
'''
@jax.jit
def findBestTh(X, y, dim, w):
    labels = jax.nn.one_hot(y, num_classes=12).squeeze() # dim n x 12
    C0 =  zeroOneLoss((w[:,None]*labels).mean(axis=0).argmax(), y, w)
    #print("C0: {} (pred = {})".format(C0, (w[:,None]*labels).mean(axis=0).argmax()))

    th = jnp.linspace(X[:,dim].min(), X[:,dim].max(), 8)
    #print("nb_th: {}".format(len(th)))
    ind = (X[:,dim, None]<th[None,:]) # dim n x n-1
    labels = labels[:,None,:] # dim n x 1 x 12
    ind2 = ind[:,:,None] # dim n x n-1 x 1
    y = y[:, None] # dim n x 1
    w1 = w[:,None] # dim n x 1
    w2 = w[:,None,None] # dim n x 1 x 1

    # left
    y_pred_l = (labels * (ind2*w2)).sum(axis=0).argmax(axis=1) # de dim n-1 puisqu'on a réduit sur 0 et 2 (devenu 1)
    Cl = zeroOneLoss(y_pred_l[None, :], y, weights=(ind)*w1)
    nl = (w1*ind).sum(axis=0) # nb valides x nb à gauche
    #print("yl: {} Cl: {} nl: {}".format(y_pred_l, Cl, nl))
    # right
    y_pred_r = (labels * (~ind2)*w2).sum(axis=0).argmax(axis=1) # de dim n-1 puisqu'on a réduit sur 0 et 2 (devenu 1)
    Cr =  zeroOneLoss(y_pred_r[None, :], y, weights=(~ind)*w1)
    nr = (w1*~ind).sum(axis=0) # nb valides x nb à droite
    #print("yr: {} Cr: {} nr: {}".format(y_pred_r, Cr, nr))

    C = C0 - (nl*Cl + nr*Cr)/(1e-12+nl+nr)
    
    m = C.argmax()
    #print('m: {} G: {} C0: {} C: {} Cl: {} Cr: {} nl: {} nr: {}'.format(m, C.max(), C0, C, Cl, Cr, nl, nr))
    return C[m], th[m]
```

```python
%%time 
G, th = findBestTh(X_train, y_train, 150, jnp.ones(len(y_train)))
print(G, th)
left = X_train[:,150] <= th
righ = X_train[:,150] > th
#left_pred = (y_train*left).mean() > 0.5
left_pred = (jax.nn.one_hot(y_train, num_classes=12).squeeze()*left[:, None]).mean(axis=0).argmax()
print("lpred: {}".format(left_pred))
#righ_pred = (y_train*righ).mean() > 0.5
righ_pred = (jax.nn.one_hot(y_train, num_classes=12).squeeze()*righ[:,None]).mean(axis=0).argmax()
print("rpred: {}".format(righ_pred))
y_pred = left_pred*left + righ_pred*righ
print(y_pred.shape)
print(zeroOneLoss(y_pred, y_train, jnp.ones(len(y_train))))
print(zeroOneLoss(jnp.ones(len(y_train)), y_train, jnp.ones(len(y_train))))
print(zeroOneLoss(jnp.zeros(len(y_train)), y_train, jnp.ones(len(y_train))))
```

We can vectorize over the dimension by using vmap. The batched function can now operate on a array of dimensions.

```python
batched_findBestTh = jax.vmap(findBestTh, in_axes=(None, None, 0, None), out_axes=0)
```

<div class="alert alert-success"> 
    <b>Q2.</b> Wrap the batched function in a function that test all dimensions to find the best combination of component and threshold. Use blocks of 256 dimensions to process at a time, as we found it a good setup with respect to speed (You can change those values later to optimize for speed).Test it on the entire train set and make sure it obtains the lowest error.<i>(Indicative time: It could take you 15 minutes to an hour to code and should run in less than 5 seconds.)</i>
</div>

```python
@jax.jit
def findBestDTh(X, y, w):
    gains = jnp.zeros((0,))
    thres = jnp.zeros((0,))
    batch = 256
    _, nbd = X.shape
    for i in range(nbd//batch+1):
        g, t = batched_findBestTh(X, y, jnp.arange(i*batch,min(nbd, (i+1)*batch)), w)
        gains = jnp.concatenate([gains, g])
        thres = jnp.concatenate([thres, t])
    
    d = gains.argmax()
    best_gain = gains[d]    
    best_dim = d
    best_th = thres[d]
    return best_gain, best_dim, best_th
```

```python
%%time
g, d, t = findBestDTh(X_train, y_train, jnp.ones(len(X_train)))
print("gain: {} dim: {} th: {}".format(g, d, t))
```

<div class="alert alert-success"> 
    <b>Q3.</b> Implement the code of the Decision Tree class by using the previous functions. To achieve reasonable speed, loop only over dimensions that have variations (there is no threshold if all samples have the same value) in batches of 256 or more, inspired by the previous function. Do not split and slice the data but zero the associated weights instead, it is faster (all functions have the same size of arrays and are thus compiled and optimized only once). Debug it on only 256 dimensions, because using all dimensions takes about 2 minutes. Test it on the full set with a maximum depth of 8 and a leaf size less than 10 to analyze and comment. <i>(Indicative time: It could take you 30 minutes to an hour to code and debug since it involves recursion.)</i>
</div>

```python
class RandomizedDT():
    def __init__(self, percent_dimension=1.0, max_depth=8, max_size=20, verbose=False, space=0):
        self.percent_dimension = percent_dimension
        self.max_depth = max_depth
        if max_depth > 0:
            self.left = RandomizedDT(percent_dimension, max_depth=max_depth-1, max_size=max_size, verbose=verbose, space=space+2)
            self.right = RandomizedDT(percent_dimension, max_depth=max_depth-1, max_size=max_size, verbose=verbose, space=space+2)
        self.label = None
        self.max_size = max_size
        self.verbose = verbose
        self.space = " "*space
    
    '''
    train this decision tree on a random subset of dimensions (columns) of X with a maximum depth
    '''
    def fit(self, X, y, w=None):
        X = jnp.array(X)
        y = jnp.array(y)
        if w is None:
            w = jnp.ones(len(X))
        if self.verbose:
            print(self.space + 'max_depth {} X: {} y: {} w: {}'.format(self.max_depth, X.shape, y.shape, w.sum()))
        label = jax.nn.one_hot(y, num_classes=12)
        
        d = X.shape[1]
        r = jnp.array(np.random.permutation(d)[0:int(d*self.percent_dimension)])

        s = (X[w>0][:, r]).std(axis=0)
        r = r[s>0]
        
        if self.max_depth <= 0 or w.sum() <= self.max_size or (w[:,None]*label).sum(axis=0).max() == w.sum() or len(r) == 0:
            if self.verbose:
                print(self.space + 'found label: {} {}'.format((w[:,None]*label).sum(axis=0), w.sum()))
            self.label = (w[:, None]*label).mean(axis=0).argmax()
            return
        if self.verbose:
            print(self.space+'looping though {} dimensions'.format(len(r)))
        
        gains = jnp.zeros((0,))
        thres = jnp.zeros((0,))
        batch = 256
        for i in range(len(r)//256+1):
            g, t = batched_findBestTh(X, y, r[i*batch:(i+1)*batch], w)
            gains = jnp.concatenate([gains, g])
            thres = jnp.concatenate([thres, t])
        
        d = gains.argmax()
        best_gain = gains[d]
        
        if best_gain <= 0:
            self.label = (w[:, None]*label).mean(axis=0).argmax()
            return
        
        self.label = None
        best_dim = r[d]
        self.dim = best_dim
        best_th = thres[d]
        self.th = best_th
        if self.verbose:
            print(self.space + 'found dim: {} theta: {} gain: {}'.format(best_dim, best_th, best_gain))
        ind = X[:, best_dim] < best_th
        if self.verbose:
            print(self.space + 'fitting left')
        self.left.fit(X[ind, :], y[ind], w[ind])
        if self.verbose:
            print(self.space + 'fitting right')
        self.right.fit(X[~ind], y[~ind], w[~ind])
        return       
    
    '''
    predict the set of samples
    '''
    def predict(self, X):
        if self.label is not None:
            return self.label * jnp.ones(len(X))
        return jnp.concatenate([self.left.predict([x]) if x[self.dim] < self.th else self.right.predict([x]) for x in X])
```

## Random Forest

Next, we want to mitigate the tendancy of decision trees to overfit when the depth is too high and to underfit when the depth is too small by implementing random forests.

<div class="alert alert-success"> 
    <b>Q5.</b> Code a Random Forest of decision trees, each trained on a subset of the training set. Perform a corase cross-validation to set a reasonnable number of trees (25, 50, 75), percent of training data used (0.5, 0.75), percent of dimensions used (0.5, 0.75) and depth 3. <i>(Indicative time: less than 20 minutes to code, takes more than 20 minutes to run)</i>
</div>

```python
class RandomForest():
    def __init__(self, nb_trees, percent_dataset=1., percent_dimension=1., max_depth=8):
        self.nb_trees = nb_trees
        self.percent_dataset = percent_dataset
        self.percent_dimension = percent_dimension
        self.max_depth = max_depth
        self.trees = []
        
    def fit(self, X, y):
        for t in range(self.nb_trees):
            dt = RandomizedDT(percent_dimension=self.percent_dimension, max_depth=self.max_depth)
            r = np.random.permutation(len(X))[0:int(self.percent_dataset*len(X))]
            dt.fit(X[r,:], y[r])
            self.trees.append(dt)
        return
    
    def predict(self, X):
        y = []
        for dt in self.trees:
            y.append(dt.predict(X))
        y = jax.nn.one_hot(jnp.array(y), num_classes=12)
        return y.sum(axis=0).argmax(axis=1)
```

```python
%%time

for n in range(25, 100, 25):
    for t in [0.5, 0.75]:
        for d in [0.5, 0.75]:
            err = []
            for p in range(3):
                r = np.random.permutation(len(y_train))
                x_t = X_train[r[0:600], :]
                y_t = y_train[r[0:600]]
                x_v = X_train[r[600:1200], :]
                y_v = y_train[r[600:1200]]
                dt = RandomForest(nb_trees=n, percent_dataset=t, percent_dimension=d, max_depth=6)
                dt.fit(x_t, y_t)
                err.append(zeroOneLoss(dt.predict(x_v), y_v, jnp.ones(len(y_v))))
                print(n, t, d)
            err = jnp.array(err)
            print('trees: {} training: {} dimension: {} mean: {} stddev: {}'.format(n, t, d, err.mean(), err.std()))
```
